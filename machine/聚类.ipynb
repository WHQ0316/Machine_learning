```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
def draw_result(train_x, labels, cents, title):
    n_cluster = np.unique(labels).shape[0]
    color = ['red', 'blue', 'green','yellow']
    plt.figure()
    plt.title(title)
    for i in range(n_cluster):
        current_data = train_x[labels == i]
        plt.scatter(current_data[:, 0], current_data[:,1], color=color[i])
        plt.scatter(cents[i,0], cents[i,1],c='yellow',marker='*',s=100)
    return plt
```

```python
iris = datasets.load_iris()
X = iris.data
clf = KMeans(n_clusters=3)
clf.fit(X)
draw_result(X,clf.labels_,clf.cluster_centers_,'Kmeans').show()
```
