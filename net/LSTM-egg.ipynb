# LSTM --输入输出

```python
import numpy
import torch
import torch.nn as nn
```

```python
lstm = nn.LSTM(input_size=1,hidden_size=10,num_layers=1,batch_first=True)
            #    输入维度        隐藏层维度      隐藏层个数
```

```python
x = torch.randn(3,8,1) # [batch_size,squ_len,input_size]
```

```python
print(x)
print(x.shape)
```

```python
h0 = torch.randn(1,3,10)  # [num_layer,batch_size,hidden_size]
c0 = torch.randn(1,3,10)  # [num_layer,batch_size,hidden_size]
```

```python
print(h0)
print(c0)
```

```python
out,(hn,cn)= lstm(x,(h0,c0))
```

```python
print(out)
```

```python
print(out.shape)
```

```python
fc1 = nn.Linear(in_features=10,out_features=1)
y = fc1(out)
```

```python
print(y.shape)
```

```python
print(y[:,-1,:])  #取每批数据的最后一个作为每批得到的数据
```
